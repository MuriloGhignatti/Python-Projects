### Vetores especiais

Vetores especiais.

$${\mathbf i} = (1, 0, 0)$$
$${\mathbf j} = (0, 1, 0)$$
$${\mathbf k} = (0, 0, 1)$$

Deste modo, um dado vetor pode ser expresso por
$${\mathbf a} = a_x {\mathbf i} + a_y {\mathbf j} + a_z {\mathbf k}$$
em que $a_x$, $a_y$, $a_z$ são as *componentes* do vetor ${\mathbf a}$.

As principais propriedades de um vetor são o *comprimento* e a *direção*. O comprimento de um vetor ${\mathbf a}$, denotado por $|{\mathbf a}|$ é um *escalar* definido por aplicação do *Teorema de Pitágoras* a um sólido retangular.
$$|{\mathbf a}| = \sqrt{a_x^2 + a_y^2 + a_z^2}$$

# Produtos de Vetores

Um *vetor unitário* é qualquer vetor de comprimento igual a *um*, independentemente da direção. Um *vetor unitário* na direção de ${\mathbf a}$ é denotado por ${\mathbf â}$, em que

$$ {\mathbf â} = \frac{\mathbf a}{|{\mathbf a}|} $$

Ou seja,

$${\mathbf â} = (\frac{a_x}{|{\mathbf a}|},\ \frac{a_y}{|{\mathbf a}|},\ \frac{a_z}{|{\mathbf a}|})$$

As componentes do vetor unitário $\mathbf{â}$ podem então ser definidas por

$${\mathbf â}_x = \frac{a_x}{|{\mathbf a}|},\quad {\mathbf â}_y = \frac{a_y}{|{\mathbf a}|},\quad {\mathbf â}_z = \frac{a_z}{|{\mathbf a}|} $$

Deve-se observar que, se $\alpha$, $\beta$ e $\gamma$ são os ângulos entre ${\mathbf a}$ e os eixos $x$, $y$ e $z$ respectivamente, tem-se

$${\mathbf â}_x = \frac{a_x}{|{\mathbf a}|} = \mathrm{cos}(\alpha), \quad {\mathbf â}_y = \frac{a_y}{|{\mathbf a}|} = \mathrm{cos}(\beta), \quad {\mathbf â}_z = \frac{a_z}{|{\mathbf a}|} = \mathrm{cos}(\gamma)$$

### Produto Escalar de Vetores

Os vetores $\mathbf a$ e $\mathbf b$ podem ser multiplicados de dois diferentes modos.

O primeiro modo produz como resultado um valor *escalar* (e não um vetor), nosso caso um *número real*.

$${\mathbf a} \cdot {\mathbf b} = a_x b_x + a_y b_y + a_z b_z $$

O *produto escalar* muitas vezes é também chamado de *dot product*.

Observar que o produto escalar é *comutativo*:$\quad\quad$  ${\mathbf a} \cdot {\mathbf b} = {\mathbf b} \cdot {\mathbf a}$

É fácil mostrar que

$${\mathbf a} \cdot {\mathbf a} = |{\mathbf a}|^2 $$ 

Pode-se também mostrar que o ângulo entre dois vetores, $\theta$, satisfaz a equação

$${\mathbf a} \cdot {\mathbf b} = |{\mathbf a}|\ |{\mathbf b}|\ {\textrm cos}(\theta)$$

Se ${\textrm cos}(\theta) = 0$, temos então que os vetores $\quad$ ${\mathbf a}$ e ${\mathbf b}$ $\quad$ são *ortogonais*, isto é, *perpendiculares* entre si.

Se $\theta = 0$, temos então que os vetores $\quad$ ${\mathbf a}$ e ${\mathbf b}$ $\quad$ são *paralelos*.

#### Propriedades do Produto Escalar

1. ${\mathbf a} \cdot {\mathbf b} = |{\mathbf a}|\ |{\mathbf b}|\ \textrm{cos}(\theta)$
1. ${\mathbf a} \cdot {\mathbf a} = |{\mathbf a}|^2$
1. ${\mathbf a} \cdot {\mathbf b} = {\mathbf b} \cdot {\mathbf a}$, propriedade comutativa
1. ${\mathbf a} \cdot ({\mathbf b} + {\mathbf c}) = {\mathbf a} \cdot {\mathbf b} + {\mathbf a} \cdot {\mathbf c}$, propriedade distributiva
1. $(k {\mathbf a}) \cdot {\mathbf b} = {\mathbf a} \cdot (k {\mathbf b}) = k ({\mathbf a} \cdot {\mathbf b})$, propriedade associativa
1. Se ${\mathbf a}$ é perpendicular a ${\mathbf b}$, então ${\mathbf a} \cdot {\mathbf b} = 0$

### Produto Vetorial

O *produto vetorial* de dois vetores ${\mathbf a}$ e ${\mathbf b}$ é também um vetor

$${\mathbf a} \times {\mathbf b}\quad =\quad (a_y b_z - a_z b_y,\quad -[a_x b_z - a_z b_x],\quad a_x b_y - a_y b_x) $$

Se ${\mathbf c} = {\mathbf a} \times {\mathbf b}$ então ${\mathbf c}$ é perpendicular a ${\mathbf a}$ e a ${\mathbf b}$, portanto é perpendicular ao plano definido por ${\mathbf a}$ e ${\mathbf b}$.

O produto vetorial *não é comutativo*

$${\mathbf b} \times {\mathbf a} = -({\mathbf a} \times {\mathbf b})$$

```python
def somaVetores(v1:tuple, v2:tuple):
  if not(type(v1)==tuple and type(v2)==tuple):
    return "Vetores devem ser do tipo tuple"
  if not(len(v1) == len(v2)):
    return "Vetores devem ter o mesmo comprimento"
  return tuple([v1[i]+v2[i] for i in range(len(v1)) ])

def multEscalarVetor(k:float, v:tuple):
  return tuple([k*elemento for elemento in v])

equacaoReta = lambda p0: lambda p1: lambda u: somaVetores(p0, multEscalarVetor(u, somaVetores(p1, multEscalarVetor(-1., p0))))
```

A combinação de três cores é denominada *pixel*.

Produção de imagem em tons de cinza.

```python
def rgb2gray(imrgb):
  fil = [0.299, 0.587, 0.144, 1]
  return np.dot(imrgb, fil)
```

```python
img2 = rgb2gray(img1)
py.imshow(img2,cmap='gray')
```

Aplicação de filtros à imagem

```python
import scipy.signal as scs

filtered = scs.medfilt(img1)
py.imshow(filtered)
```

```python
filtered = scs.wiener(img1)
py.imshow(filtered)
```

Seleção de pontos segundo a tonalidade.

```python
reddish = filtered[:, :, 0] > 180/255 # reddish é uma matriz booleana
fx = filtered.copy() # copiar um array (sem alterar o original)
fx[reddish] = [1, 0, 0, 1] # os pts True de fx são alterados
py.imshow(fx)
```

```python
from scipy import ndimage
# ndimage: módulo para rotação de imagens
```

```python
# ângulo de rotação: 10 graus
# mode: modo como os pixels situados fora do array original são tratados
# cval: valor atribuído aos pixels situados fora do array original
im_rot = ndimage.rotate(img,25,mode='constant',cval=0.8)

# método de interpolação: nearest
py.imshow(im_rot,interpolation='nearest')
```
