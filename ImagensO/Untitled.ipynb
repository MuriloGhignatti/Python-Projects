```python
import numpy as np
import matplotlib.pyplot as py
```

```python
img1 = py.imread("forest.png")
img2 = py.imread("otherforest.jpg")
```

```python
def atribute(img1,img2,SideBySide):
    M0 = 0
    M1 = 0
    M2 = 0
    if img1.shape[0] > img2.shape[0]:
        M0 = img2.shape[0]
    else:
        M0 = img1.shape[0]
    if img1.shape[1] > img2.shape[1]:
        M1 = img2.shape[1]
    else:
        M1 = img1.shape[1]
    if img1.shape[2] != img2.shape[2]:
        return "As Imagens Devem Ter o Mesmo Numero de Camadas"
    else:
        M2 = img1.shape[2]
    if SideBySide:
        return np.zeros((M0*2,M1*2,M2))
    else:
        return np.zeros((M0,M1,M2))
```

```python
def SideBySide(img1,img2):
    x = 0
    Result  = atribute(img1,img2,True)
    items2 = [img2[x][y] for x in range(int(Result.shape[0]/2)) for y in range(int(Result.shape[1]/2))]
    for L in range(0,int(Result.shape[0]/2)):
        for C in range(0,int(Result.shape[1]/2)):
            Result[L][C] = img1[L][C]
    for l in range(int(Result.shape[0]/2),Result.shape[0]-1):
        for c in range(int(Result.shape[1]/2),Result.shape[1]-1):
            x+=1
            Result[l][C] = items2[x]
    return Result,items2.__len__()
```

```python
img3 = SideBySide(img1,img2)
```
